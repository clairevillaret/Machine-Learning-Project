```python
music_preferences = df.iloc[:, :19]
movie_preferences = df.iloc[:, 19:31]
hobbiesAnd_interests = df.iloc[:, 31:63]
phobias = df.iloc[:, 63:73]
health_habits = df.iloc[:, 73:76]
traits_views_opinions = df.iloc[:, 76:133]
spending_habits = df.iloc[:, 133:140]
demographics = df.iloc[:, 140:150]
hobbiesAnd_interests
```

```python
#remove music and movie preferences from dataset
music_preferences = df.iloc[:, :19]
movie_preferences = df.iloc[:, 19:31]
df_copy = df.copy()
# df.drop(music_preferences, inplace=True, axis=1)
# df.drop(movie_preferences, inplace=True, axis=1)
df_copy.drop(music_preferences, inplace=True, axis=1)
df_copy.drop(movie_preferences, inplace=True, axis=1)
df_copy
```

```python
#Find the missing data in the columns
df_copy.isnull().sum()
```

```python
#Filling missing data
df_copy.fillna(0,axis=1,inplace=True)
df_copy.isnull().sum()
```

```python
#remove the demographics from dataset because we only want to focus on lifestyle behaviours
data = df_copy.iloc[:,:109]
data
```

```python
#convert strings to numerical categories
columnList = ['Internet usage', 'Alcohol','Smoking']
data[columnList] = data[columnList].apply(lambda x: pd.factorize(x)[0])
#delete unwanted variables
data.pop('Lying')
data.pop('Punctuality')
data
```

```python
column_to_move = data.pop('Loneliness')
data.insert(106, 'Loneliness', column_to_move)
data
```

```python
#separate the dependent variable and independent variables
x = data.iloc[:,:106]
y = data.iloc[:,106:]  #loneliness variable is the dependent variable
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split dataset into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
f_score = chi2(x_train, y_train)
f_score[1]
```

```python
pvalues = pd.Series(f_score[1])
pvalues.index = x_train.columns
pvalues
```

```python
pvalues.sort_values(ascending=False)
```

```python
#select the variables that have p-value of less than 0.05 because it means they are significant and can be used as variable to the model
pvalues.head()
```

```python
features = []
list1 = []

features = list(pvalues[pvalues<0.05].index)

for i in pvalues:
    if i < 0.05:
        list1.append(i)

final_list = pd.Series(list1, index=features)
final_list
```
